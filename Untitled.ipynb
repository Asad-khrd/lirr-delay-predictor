```python
import pandas as pd

# Load your LIRR dataset
# IMPORTANT: Replace 'lirr_data.csv' with the actual name of your file.
df = pd.read_csv('MTA_LIRR_Delays.csv')
```

```python
# Get a technical summary of the DataFrame
df.info()
```

```python
# Get a summary of all 'object' (text) columns
df.describe(include='object')
```

```python
# Define the columns we want to inspect
categorical_columns = ['Branch', 'Period', 'Status', 'Delay Category']

# Loop through each column and print its unique values and their counts
for col in categorical_columns:
    print(f"--- Values for {col} ---")
    print(df[col].value_counts())
    print("\\n" + "="*30 + "\\n") # Adds a separator for readability
```

```python
# 1. Handle Missing Values
df['Minutes Late'] = df['Minutes Late'].fillna(0)
df['Delay Category'] = df['Delay Category'].fillna('Unknown')

# 2. Convert the 'Depart Time' and 'Arrive Time' columns directly to datetime objects
# We will also rename them to be more descriptive
df['Depart DateTime'] = pd.to_datetime(df['Depart Time'])
df['Arrive DateTime'] = pd.to_datetime(df['Arrive Time'])

# 3. Drop the original, now redundant, columns for a cleaner DataFrame
# We can drop all three original date/time columns
df.drop(['Service Date', 'Depart Time', 'Arrive Time'], axis=1, inplace=True)

# 4. Verify all changes
print("--- Data Cleaned ---")
df.info()
```

```python
# Group by branch and count the number of incidents, then sort
branch_incidents = df.groupby('Branch').size().sort_values(ascending=False)

# Display the results
print(branch_incidents)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style and size for the plot
sns.set_style("whitegrid")
plt.figure(figsize=(12, 8))

# Create the bar plot using Seaborn
sns.barplot(x=branch_incidents.values, y=branch_incidents.index, palette="plasma")

# Add a title and labels for clarity
plt.title('Total LIRR Incidents (Delays & Cancellations) by Branch', fontsize=16)
plt.xlabel('Total Number of Incidents', fontsize=12)
plt.ylabel('LIRR Branch', fontsize=12)

# Display the plot
plt.show()
```

```python
# Filter for only 'Late' trains
late_trains_df = df[df['Status'] == 'Late']

# Group by branch and calculate the average delay time
avg_delay_by_branch = late_trains_df.groupby('Branch')['Minutes Late'].mean().sort_values(ascending=False)

# Display the results
print(avg_delay_by_branch)
```

```python
# --- Analysis for Babylon Branch (Most Frequent Incidents) ---

# Filter the DataFrame for only the Babylon branch
babylon_df = df[df['Branch'] == 'Babylon']

# Get the value counts for delay categories on the Babylon branch
babylon_causes = babylon_df['Delay Category'].value_counts()

# Print the top 3 causes
print("--- Delay Causes for Babylon Branch ---")
print(babylon_causes)
print("\\n" + "="*40 + "\\n") # Adds a separator

# --- Analysis for Montauk Branch (Longest Average Delays) ---

# Filter the DataFrame for only the Montauk branch
montauk_df = df[df['Branch'] == 'Montauk']

# Get the value counts for delay categories on the Montauk branch
montauk_causes = montauk_df['Delay Category'].value_counts()

# Print the top 3 causes
print("--- Delay Causes for Montauk Branch ---")
print(montauk_causes)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# --- Step 1: Create a Refined Feature Set ---
y = df['Delay Category']

# Using our most robust features, excluding the overly-specific 'Depart Station'
features = ['Branch', 'Period', 'Depart DateTime']
X = df[features].copy()

X['day_of_week'] = X['Depart DateTime'].dt.dayofweek
X['hour'] = X['Depart DateTime'].dt.hour
X = X.drop('Depart DateTime', axis=1)

# --- Step 2: Convert Categorical Features to Numbers ---
X_encoded = pd.get_dummies(X, columns=['Branch', 'Period'])

# --- Step 3: Split the Data ---
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# --- Step 4: Create and Train the TUNED Model ---
# We've added parameters to help the model generalize better
model = RandomForestClassifier(n_estimators=100,      # Number of trees
                               max_depth=20,          # Max questions per tree
                               min_samples_leaf=5,    # Min samples to make a conclusion
                               random_state=42, 
                               n_jobs=-1, 
                               class_weight='balanced')

print("Training the tuned model...")
model.fit(X_train, y_train)
print("Model training complete.")

# --- Step 5: Test the New Accuracy ---
predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)

print(f"\\nTuned Model Accuracy: {accuracy:.2%}")
```

```python
import pandas as pd

# Use our trained model to predict the probabilities for the entire test set
probabilities = model.predict_proba(X_test)

# --- Let's analyze a single example (the first row of our test data) ---
example_index = 0
example_probabilities = probabilities[example_index]

# Create a DataFrame to easily see the results
prob_df = pd.DataFrame({
    'Category': model.classes_,
    'Probability': example_probabilities
}).sort_values(by='Probability', ascending=False) # Sort to find the highest probabilities

# Display the top 3 most likely causes for this specific incident
print("--- Top 3 Likely Delay Causes for a Sample Incident ---")
print(prob_df.head(3).to_string(index=False))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a crosstab to count the occurrences of each category in each period
crosstab = pd.crosstab(df['Period'], df['Delay Category'])

# Normalize to get the proportion (percentage) across each row
crosstab_normalized = crosstab.div(crosstab.sum(axis=1), axis=0)

# Create the stacked bar chart
crosstab_normalized.plot(kind='bar', 
                         stacked=True, 
                         figsize=(14, 8), 
                         colormap='viridis')

plt.title('Proportion of Delay Causes by Period of Day', fontsize=16)
plt.ylabel('Proportion', fontsize=12)
plt.xlabel('Period of Day', fontsize=12)
plt.xticks(rotation=0) # Keep the x-axis labels horizontal
plt.legend(title='Delay Category', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# --- Step 1: Create a Feature Set with ALL Contextual Features ---
y = df['Delay Category']

# Including all potentially useful categorical and time-based features
features = ['Train', 'Branch', 'Period', 'Depart Station', 'Depart DateTime']
X = df[features].copy()

X['day_of_week'] = X['Depart DateTime'].dt.dayofweek
X['hour'] = X['Depart DateTime'].dt.hour
X = X.drop('Depart DateTime', axis=1)

# --- Step 2: Convert All Categorical Features to Numbers ---
X_encoded = pd.get_dummies(X, columns=['Train', 'Branch', 'Period', 'Depart Station'])

# --- Step 3: Split the Data ---
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# --- Step 4: Create and Train the Tuned Model on All Features ---
model = RandomForestClassifier(n_estimators=100,
                               max_depth=20,
                               min_samples_leaf=5,
                               random_state=42, 
                               n_jobs=-1, 
                               class_weight='balanced')

print("Training the final model with all features... (This will be slow)")
model.fit(X_train, y_train)
print("Model training complete.")

# --- Step 5: Test the Final Accuracy ---
predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)

print(f"\\nFinal Model Accuracy (with all features): {accuracy:.2%}")
```

```python
# --- Create the 'rush_hour' feature for EDA ---
def get_rush_hour(hour):
    if 6 <= hour <= 9: return 'AM_Rush'
    elif 16 <= hour <= 19: return 'PM_Rush'
    else: return 'Off_Peak'
df['rush_hour'] = df['Depart DateTime'].dt.hour.apply(get_rush_hour)

# --- Create and plot the comparison ---
crosstab_rush = pd.crosstab(df['rush_hour'], df['Delay Category'])
crosstab_rush_normalized = crosstab_rush.div(crosstab_rush.sum(axis=1), axis=0)

crosstab_rush_normalized.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='plasma')
plt.title('Comparison of Delay Causes by Rush Hour', fontsize=16)
plt.ylabel('Proportion of Delays')
plt.xlabel('Time of Day')
plt.xticks(rotation=0)
plt.legend(title='Delay Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# --- Clean up the temporary column ---
df.drop('rush_hour', axis=1, inplace=True)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import holidays

# --- Step 1: Create All Validated Features ---
# Hub Station Feature
top_10_stations = df['Depart Station'].value_counts().head(10).index
df['is_hub_station'] = df['Depart Station'].apply(lambda x: 1 if x in top_10_stations else 0)

# Holiday Feature
us_holidays = holidays.US()
df['is_holiday'] = df['Depart DateTime'].dt.date.apply(lambda x: 1 if x in us_holidays else 0)

# Rush Hour Feature
def get_rush_hour(hour):
    if 6 <= hour <= 9: return 'AM_Rush'
    elif 16 <= hour <= 19: return 'PM_Rush'
    else: return 'Off_Peak'
df['rush_hour'] = df['Depart DateTime'].dt.hour.apply(get_rush_hour)


# --- Step 2: Create the Final, Most Powerful Feature Set ---
y = df['Delay Category']

features = ['Branch', 'rush_hour', 'is_holiday', 'is_hub_station', 'Depart DateTime']
X = df[features].copy()

X['day_of_week'] = X['Depart DateTime'].dt.dayofweek
X['hour'] = X['Depart DateTime'].dt.hour
X = X.drop('Depart DateTime', axis=1)

# --- Step 3: One-Hot Encode and Split ---
X_encoded = pd.get_dummies(X, columns=['Branch', 'rush_hour'])
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# --- Step 4: Train and Test the Definitive Model ---
model = RandomForestClassifier(n_estimators=100,
                               max_depth=20,
                               min_samples_leaf=5,
                               random_state=42,
                               n_jobs=-1,
                               class_weight='balanced')

print("Training the definitive model with all validated features...")
model.fit(X_train, y_train)
print("Model training complete.")

predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)

print(f"\\nDefinitive Model Accuracy: {accuracy:.2%}")
```
